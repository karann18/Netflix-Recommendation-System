```python
df['combined_features'] = df['title'] + " " + df['director'] + " " + df['cast'] + " " + df['listed_in'] + " " + df['description']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

#  Here Initializing  TF-IDF Vectorizer
tfidf = TfidfVectorizer(stop_words='english')

# Here Transforming the combined text column into vectors
tfidf_matrix = tfidf.fit_transform(df['combined_features'])

# Here Checking the shape of the matrix
print(tfidf_matrix.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Here Computing the cosine similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix)

# Check the shape
print(cosine_sim.shape)
```

```python
# Create a Series to map movie/show titles to their index
indices = pd.Series(df.index, index=df['title']).drop_duplicates()

def recommend(title, cosine_sim=cosine_sim):
    if title not in indices:
        return "Title not found in the dataset. Try another one."

    # Get the index of the movie/show
    idx = indices[title]

    # Get similarity scores of all shows with the selected one
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort movies/shows by similarity score (highest first)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get indices of the top 10 most similar shows (excluding itself)
    sim_indices = [i[0] for i in sim_scores[1:11]]

    # Return the top 10 recommended titles
    return df['title'].iloc[sim_indices]

# Test the function
recommend("Breaking Bad")  # Replace with any title from your dataset
```

```python
recommend("Zombieland") 
```

```python
while True:
    user_input = input("Enter a movie/show title (or type 'exit' to stop): ")
    if user_input.lower() == "exit":
        break
    print("\nRecommended Shows/Movies:")
    print(recommend(user_input))
    print("\n" + "-"*50 + "\n")
```
